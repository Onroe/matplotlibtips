```python
import matplotlib.pyplot as plt
import matplotlib.image as image

def bar_chart(numbers, labels, pos):
    fig = plt.figure(figsize=(10, 10))
    plt.bar(pos,numbers, color="green")
    
    #add watermark
    fig.text(0.5, 0.2, '@ Inc', fontsize=45, color='blue', ha='right', va='bottom', alpha=0.4, rotation=25)
    plt.xticks(pos,labels)   
    
    #add footnote
    plt.annotate('Designed by Inc', xy = (1.0, -0.05),xycoords='axes fraction',ha='right',va="center",fontsize=8)
    
    # fetch logo and add logo
    logo = image.imread('logo.jpg')
    newax = fig.add_axes([0.6,0.6,0.1,0.1], anchor='NE', zorder=0)
    newax.imshow(logo)
    newax.axis('off')
    plt.show()
    

if __name__ == '__main__':
    numbers= [40,50,22,30]
    labels= ['UGX', 'USD', 'Pound', 'EURO'] 
    pos = list(range(4))
    bar_chart(numbers, labels,pos)
   
```
